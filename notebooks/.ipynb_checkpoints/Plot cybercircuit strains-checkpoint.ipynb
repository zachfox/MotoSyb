```python
%matplotlib notebook 

from importlib import reload
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import dnaplotlib
reload(dnaplotlib)
import matplotlib.pyplot as plt
from matplotlib import gridspec
# more parts 
tet_color = [41/255,171/255,226/255]
cI_color = [247/255,147/255,30/255]
lac_color = [57/255,181/255,74/255]

pccas = {'name':'p_ccas', 'start':1, 'end':10, 'type':'Promoter', 'opts': {'color':[0.0, 0.0, 0.0]}}
tetr = {'name':'tetR', 'start':11, 'end':30, 'type':'CDS',
        'opts':{'label': 'tetR', 'fontsize': 8,  'label_y_offset': 0, 
                'label_x_offset': -2, 'label_style':'italic', 'color':tet_color},'model_type':'transcription_translation'}

ptet = {'name':'P_tet', 'start':31, 'end':41, 'type':'Promoter',
        'opts': {'color':tet_color}}

laci = {'name':'lacI', 'start':76, 'end':95, 'type':'CDS', 'opts':{'label': 'lacI', 'fontsize': 8,  'label_y_offset': 0, 'label_x_offset': -2, 'label_style':'italic', 'color':lac_color}}
plac = {'name':'P_lac', 'start':1, 'end':10, 'type':'Promoter', 'opts': {'color':lac_color}}


yfp = {'name':'yfp', 'start':42, 'end':52, 'type':'CDS', 
       'opts':{'label': 'yfp', 'fontsize': 8,  'label_y_offset': 0,
               'label_x_offset': -2, 'label_style':'italic', 'color':[0.38, 0.82, 0.32]},'model_type':'transcription_translation'}

cI = {'name':'cI', 'start':131, 'end':150, 'type':'CDS', 'opts':{'label': 'cI', 'fontsize': 8, 'label_y_offset': 0, 'label_x_offset': -1, 'label_style':'italic', 'color':cI_color}}
pcI = {'name':'P_cI', 'start':56, 'end':65, 'type':'Promoter', 'opts': {'color':cI_color}}


tet_repress = {'from_part':tetr, 'to_part':ptet, 'type':'Repression',
               'opts':{'linewidth':1, 'color':tet_color}}
pccas_control = {'from_part':pccas,'to_part':pccas,'type':'Control',
                 'opts':{'linewidth':1, 'color':[0, 0, 0]}}



all_parts = [pccas,tetr,ptet,yfp]
all_interactions = [tet_repress,pccas_control]
# make the plot 
f,ax = plt.subplots(figsize=(5,2))
dnaplotlib.plot_sbol_designs([ax], [all_parts],
[all_interactions])
f.show()
```

# Generate all strains and some descriptions.

```python
strain_names = ['strain 21','strain 24','strain 25','strain 19','strain 20','strain 23','strain 18','strain 26']
all_strains_all_parts = [[pccas,laci,plac,yfp],[pccas,cI,pcI,laci,plac,yfp],
             [pccas,cI,pcI,yfp],[pccas,tetr,ptet,cI,pcI,yfp],
             [pccas,tetr,ptet,yfp],
             [pccas,laci,plac,tetr,ptet,yfp],
             [pccas,tetr,ptet,cI,pcI,laci,plac,yfp],
             [pccas,cI,pcI,laci,plac,tetr,ptet,yfp]]

f,ax = plt.subplots(4,2)
ax = ax.T.ravel()
for i in range(8):
    all_parts = all_strains_all_parts[i]
    all_interactions = [pccas_control]
    for j in range(1,len(all_parts)-1,2):
        all_interactions.append({'from_part':all_parts[j],'to_part':all_parts[j+1],'type':'Repression',
                                'opts':{'linewidth':1,'color':all_parts[j]['opts']['color']}})
    dnaplotlib.plot_sbol_designs([ax[i]],[all_parts],[all_interactions])
    ax[i].set_title(strain_names[i])

f.show()
f.savefig('strains_plotted.pdf')
    
```
